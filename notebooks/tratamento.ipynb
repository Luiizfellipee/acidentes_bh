#### Separando DF por Planilha

```python
dfs_limpos = {}

for nome_aba, df in base_full.items():
    temp_df = df.copy()

    temp_df.columns = [col.lower().strip() for col in temp_df.columns]

    dfs_limpos[nome_aba] = temp_df
    
```

```python
dfs_limpos['envolvidos'] = dfs_limpos['envolvidos'].rename(columns={'num_boletim' : 'cod_boletim'})
dfs_limpos['envolvidos'].head(5)
```

```python
dfs_limpos['envolvidos']['idade'] = dfs_limpos['envolvidos']['idade'].astype('Int64')
```

```python
dfs_limpos['envolvidos']['idade'].value_counts().sort_index()
```

```python
import numpy as np

# 1. Calculamos a MEDIANA apenas das idades que fazem sentido (ex: entre 1 e 98)
filtro_idades_reais = dfs_limpos['envolvidos']['idade'].between(1, 98)
mediana_idade = dfs_limpos['envolvidos']['idade'][filtro_idades_reais].median()

print(f"A mediana real calculada foi: {mediana_idade}")

# 2. Criamos uma função de limpeza
def limpar_idade(idade):
    # Se for 0 ou maior que 98, usamos a mediana
    if idade == 0 or idade > 98 or pd.isna(idade):
        return int(mediana_idade)
    return int(idade)

# 3. Aplicamos a limpeza
dfs_limpos['envolvidos']['idade'] = dfs_limpos['envolvidos']['idade'].apply(limpar_idade)

# 4. Agora convertemos para o int64 (minúsculo) - O Parquet vai amar isso!
dfs_limpos['envolvidos']['idade'] = dfs_limpos['envolvidos']['idade'].astype('int64')

# Conferindo o resultado
print(dfs_limpos['envolvidos']['idade'].value_counts().sort_index())
```

```python
# Trocando N por NÃO e S por SIM
dfs_limpos['envolvidos']['pedestre'] = dfs_limpos['envolvidos']['pedestre'].replace({'N': 'NÃO', 'S': 'SIM'})
dfs_limpos['envolvidos']['passageiro'] = dfs_limpos['envolvidos']['passageiro'].replace({'N': 'NÃO', 'S': 'SIM'})
dfs_limpos['envolvidos']['condutor'] = dfs_limpos['envolvidos']['condutor'].replace({'N': 'NÃO', 'S': 'SIM'})
```

```python
colunas_cnh = ['categoria_habilitacao', 'descricao_habilitacao', 'especie_veiculo']

for col in colunas_cnh:
    dfs_limpos['envolvidos'][col] = dfs_limpos['envolvidos'][col].fillna('NAO SE APLICA')
```

```python
dfs_limpos['envolvidos'].isnull().sum()
```

```python
conflito = dfs_limpos['envolvidos'][(dfs_limpos['envolvidos']['pedestre'] == 'SIM') & (dfs_limpos['envolvidos']['passageiro'] == 'SIM')]
print(f"Quantidade de linhas com erro de lógica: {len(conflito)}")
```

```python
col_pp = ['pedestre', 'passageiro']

for col in col_pp:
    dfs_limpos['envolvidos'][col] = dfs_limpos['envolvidos'][col].fillna('NAO INFORMADO')
```

```python
tabelas_secundarias = ['envolvidos', 'veiculos', 'logradouros']

for nome in tabelas_secundarias:
    dfs_limpos[nome] = dfs_limpos[nome].drop(columns=['data', 'hora'], errors= 'ignore')
```

```python
dfs_limpos['veiculos'] = dfs_limpos['veiculos'].rename(columns={'nº_boletim' : 'cod_boletim'})
dfs_limpos['logradouros'] = dfs_limpos['logradouros'].rename(columns={'nº_boletim' : 'cod_boletim'})
```

```python
import os
import pandas as pd
import numpy as np

folder_path = '../data/dados_tratados'
os.makedirs(folder_path, exist_ok=True)

for nome_tabela, df in dfs_limpos.items():
   
    temp_df = df.copy()

    
    if 'idade' in temp_df.columns and nome_tabela == 'envolvidos':
        
        mediana = int(temp_df[temp_df['idade'].between(1, 98)]['idade'].median())
        
        temp_df['idade'] = temp_df['idade'].apply(lambda x: mediana if x == 0 or x > 98 or pd.isna(x) else int(x))
        temp_df['idade'] = temp_df['idade'].astype(int)

    for col in temp_df.columns:
        
        temp_df[col] = pd.Series(temp_df[col].to_numpy(), name=col)

    cols_obj = temp_df.select_dtypes(include=['object', 'string']).columns
    for col in cols_obj:
        temp_df[col] = temp_df[col].astype(str)

    file_path = os.path.join(folder_path, f"{nome_tabela}.parquet")
    
    # Se o fastparquet continuar dando erro, vamos usar o pyarrow como motor
    try:
        temp_df.to_parquet(file_path, index=False, engine='pyarrow')
    except:
        temp_df.to_parquet(file_path, index=False, engine='fastparquet')
    
    print(f"✅ Sucesso total: {file_path}")
```
